```python
sentence="The Eiffel Tower is one of the most famous monuments in the world and is located in Paris, France. It was built between 1887 and 1889 for the 1889 World’s Fair (Exposition Universelle), which was held to celebrate the 100th anniversary of the French Revolution."
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
tag_elements=nltk.pos_tag(words)
```
