# Steps to Open and Use Jupyter Notebook
Follow these steps to install, launch, and use Jupyter Notebook on your system.

## 1. Install Jupyter Notebook
Use pip to install Jupyter Notebook. Open your terminal or command prompt and run the following command:

```python
# Install Jupyter Notebook using pip
!pip install notebook
```

## 2. Launch Jupyter Notebook from Terminal
After installation, start the Jupyter Notebook server by running this command in your terminal or command prompt:

```python
# Launch Jupyter Notebook
!jupyter notebook
```

## 3. Open a Notebook File
Once the Jupyter Notebook interface opens in your web browser, navigate to the folder containing your `.ipynb` file. Click on the file to open it in a new tab.

## 4. Run Code Cells
To execute code in a cell, click on the cell and press <kbd>Shift</kbd> + <kbd>Enter</kbd>. This will run the code and display the output below the cell.
