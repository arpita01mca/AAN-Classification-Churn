### Stemming ###
Stemming in natural language processing (NLP) involves reducing words to their root form or stem, 
which may not always be a valid word. For example, "arguing" and "argued" may sometimes be stemmed to "argu," 
which isn't a correct word. This is achieved primarily by removing suffixes.



```python
## Classification Problem
## Comments of a product is positive review or negative review 
## Reviews ---> eating, eat, eaten [going, gone, goes]-->go
words=["eating","eats","eaten","writing","writes","programming"]
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
stemming.stem("congratulations")
```

```python
stemming.stem("sitting")
```

```python
##RegexpStemmer Class##
The RegexpStemmer class in NLTK allows the use of regular expressions to define stemming rules. 
It removes prefixes or suffixes that match a given regular expression. In this example, the suffixes “ing,” “s,” “e,” and “able” 
are removed if the remaining stem has a minimum length of 4 characters.
    
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

### Snowball Stemmer ###

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer('english')
```

```python
for word in words:
     print(word+"----->"+snowballstemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
snowballstemmer.stem("fairly"),snowballstemmer.stem("sportingly")
```
