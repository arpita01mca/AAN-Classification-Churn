### Bag Of Words ###

```python
data = [
    ("ham", "Hey, are we meeting today"),
    ("ham", "Please call me when you are free"),
    ("spam", "Congratulations! You have won a free ticket"),
    ("spam", "Win cash prizes now, click this link"),
    ("ham", "Let us go for lunch tomorrow"),
    ("spam", "Urgent! Your account has been selected for a reward"),
]
```

```python
pip install pandas
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=["label", "message"])
print(df)
```

```python
df
```

```python
## Data Cleaning and Preprocessing
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
pip install nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-z]','',df['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=''.join(review)
    corpus.append(review)
```

```python
## Create Bag Of Words Model
from sklearn.feature_extraction.text import CountVectorizer
## For binary Bag Of Words enable binary=True
cv=CountVectorizer(max_features=2500,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
```
