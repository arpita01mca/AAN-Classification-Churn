```python
## Speech of APJ Abdul Kalam ##
paragraph=""" Today, I stand before you to speak about the Vision of India—a vision of a strong, self-reliant, inclusive, and developed nation.
India is not just a land of ancient history and rich culture; it is a nation of immense potential. With more than half of our population below the age of 35, the youth of India are the driving force of our future. The vision of India lies in empowering this youth with education, skills, and values.
A developed India is one where knowledge is power. Education must not only help us earn a living but also teach us how to think, innovate, and serve society. Science and technology should reach the villages, farmers, and small industries, ensuring that development touches every citizen.
The vision of India also includes economic strength with social harmony. Growth must be inclusive—bridging the gap between the rich and the poor, between urban and rural India. Every citizen should have access to healthcare, clean water, and equal opportunities.
A truly developed India is built on ethical leadership and responsible citizenship. Corruption, discrimination, and violence weaken the nation. Honesty, unity, and respect for diversity strengthen it. As citizens, we must do our duty before asking for our rights.
India’s vision is also a peaceful nation that contributes to global progress. We believe in “Vasudhaiva Kutumbakam”—the world is one family. Through innovation, diplomacy, and compassion, India can lead the world towards peace and sustainability.
Friends, the vision of India is not the responsibility of the government alone—it is our collective responsibility. Each student, teacher, worker, and leader must contribute through dedication and hard work.
Let us dream of an India that is educated, healthy, prosperous, and united.Let us work together to turn that dream into reality.
The future of India begins with us.Jai Hind. """
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk 
nltk.download('stopwords')
```

```python
stopwords.words('English')
```

```python
stopwords.words('german')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer= PorterStemmer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
## Apply stopwords And Filter And then Apply Stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words= [stemmer.stem(word) for word in words if word not in set(stopwords.words('English'))]
    sentences[i]=' '.join(words)#conveting all the list of words into sentences
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatize=WordNetLemmatizer()
```

```python
## Apply stopwords And Filter And then Apply Stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words= [lemmatize.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('English'))]
    sentences[i]=' '.join(words)#conveting all the list of words into sentences
```
